```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
LABEL_SIZE = 28
FONTSIZE = 32

# Your data
latency = list(np.array([239, 184, 153, 132, 114, 104, 94, 90, 85, 85, 83]) / 10)
frequency = [0.31, 0.41, 0.51, 0.61, 0.71, 0.82, 0.92, 1.02, 1.12, 1.22, 1.30]
energy_consumption = [1711, 1431, 1314, 1280, 1261, 1246, 1279, 1317, 1385, 1481, 1563]

print(latency)

# Start plotting
fig, ax1 = plt.subplots(figsize=(12, 6))

# Plot latency in orange with straight lines on the primary y-axis
ax1.plot(frequency, latency, color='orange', marker='o', label='Latency')  # 'o' denotes the marker type
ax1.set_xlabel('GPU Frequency (GHz)', fontsize=FONTSIZE)
ax1.set_ylabel('Latency (s)', color='orange', fontsize=FONTSIZE)
ax1.tick_params(axis='both', labelsize=LABEL_SIZE)

latency_ticks = np.arange(0, max(latency) + 1, 5)  # Adjust step size as needed
ax1.set_yticks(latency_ticks)

# Create a second y-axis for energy consumption
ax2 = ax1.twinx()
ax2.plot(frequency, energy_consumption, color='blue', marker='o', label='Energy Consumption')  # 'o' denotes the marker type
ax2.set_ylabel('Energy Consumption (J)', color='blue', fontsize=FONTSIZE)
ax2.tick_params(axis='y', labelsize=LABEL_SIZE)

energy_ticks = np.arange(1000, max(energy_consumption) + 1, 100)  # Adjust step size as needed
ax2.set_yticks(energy_ticks)

# plt.savefig('pre_exp_dvfs.png', dpi=300)

# Display the plot
plt.show()
```
