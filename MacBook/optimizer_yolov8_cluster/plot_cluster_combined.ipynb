```python
import json
import matplotlib.pyplot as plt
import numpy as np
import os
from KVClusterV2 import KVClusterV2
from sklearn.manifold import TSNE
```

```python
# Plot
plot_directory = 'result/plot_set'
num_clusters = 10
```

```python
colors = [
	'#FF0000',
	'#00FFFF',
	'#0000FF',
	'#00008B',
	'#ADD8E6',
	'#800080',
	'#7FFFD4',
	'#008000',
	'#FF00FF',
	'#FFC0CB',
	'#C0C0C0',
	'#FFA500',
	'#000000',
	'#800000',
]
```

```python
def load_json_file(file_path):
	try:
		with open(file_path, 'r') as file:
			data = json.load(file)
		return data
	
	except Exception as e:
		print(f"An error occurred while loading the JSON file: {e}")
		return None
```

```python
def extract_fpss(metric_list):
	return list(metric_list[list(metric_list.keys())[0]][0]['metric'].keys())
```

```python
def to_accuracy_vector(accuracy_result_seq, fpss, type='F1'):
	accuracy_vector = []
	for fps in fpss:
		accuracy_vector.append(accuracy_result_seq[fps][type])
	
	return accuracy_vector
```

```python
def plot_scatter(xs, ys, x_label, y_label, title, fig_size=(8, 8)):
	plt.figure(figsize=fig_size)
	plt.scatter(xs, ys, c=colors[0], s=30, alpha=0.5)
	plt.title(title)
	plt.xlabel(x_label)
	plt.ylabel(y_label)
	plt.show()
```

```python
def plot_scatter_label(xs, ys, labels, x_label, y_label, title, fig_size=(8, 8)):
	plt.figure(figsize=fig_size)
	for i in range(len(xs)):
		plt.scatter(xs[i], ys[i], c=colors[labels[i]], s=30, alpha=0.5)
	plt.title(title)
	plt.xlabel(x_label)
	plt.ylabel(y_label)
	plt.show()
```

```python
def filter_list_index(input_list, indices_to_remove):
	return [item for i, item in enumerate(input_list) if i not in indices_to_remove]
```

## Plot

```python
plot_filenames = sorted(os.listdir(plot_directory))
plot_video_names = sorted(list(set([f.split('_')[0] for f in plot_filenames])))
```

```python
fpss = extract_fpss(load_json_file(os.path.join(plot_directory, plot_video_names[0] + "_Accuracy_Result.json")))
```

```python
np.random.seed(42)
```

```python
cluster = KVClusterV2(num_clusters)

omv_lists = []
acc_lists = []
clip_amounts = [0]

for v in plot_video_names:
	accuracy_result = load_json_file(os.path.join(plot_directory, v + "_Accuracy_Result.json"))
	movement_result = load_json_file(os.path.join(plot_directory, v + "_Movement_Result.json"))

	for class_idx in list(accuracy_result.keys()):
		for i in range(len(accuracy_result[class_idx])):
			accuracy_vector = to_accuracy_vector(accuracy_result[class_idx][i]['metric'], fpss)

			for fps in fpss:
				movement_vector = movement_result[class_idx][i]['movement'][fps]
				if movement_vector[0] != -1:
					omv_lists.append(movement_vector)
					acc_lists.append(accuracy_vector)
	
	clip_amounts.append(len(omv_lists))
```

```python
for i in range(len(omv_lists)):
	cluster.add(omv_lists[i], acc_lists[i])
cluster.cluster_average()
```

```python
clustered_omv_list = [pair[0] for pair in cluster.kv_pairs]
clustered_omv_labels = cluster.self_tell()
assert(len(clustered_omv_list) == len(clustered_omv_labels))
```

```python
# Perform t-SNE
high_dimensional_data = np.array(clustered_omv_list)
tsne = TSNE(n_components=2, random_state=42)
low_dimensional_data = tsne.fit_transform(high_dimensional_data)
```

```python
curr_idx = 0
for i in range(len(plot_video_names)):
	video_name = plot_video_names[i]
	start = clip_amounts[i]
	end = clip_amounts[i+1]

	video_omv_cop1 = low_dimensional_data[start:end, 0]
	video_omv_cop2 = low_dimensional_data[start:end, 1]
	video_label = clustered_omv_labels[start:end]

	plot_scatter_label(video_omv_cop1, video_omv_cop2, video_label, 't-SNE Component 1', 't-SNE Component 2', video_name)
```

```python
clustered_acc_list = [pair[1] for pair in cluster.kv_pairs]

high_dimensional_omv = np.array(clustered_omv_list)
tsne_omv = TSNE(n_components=1, random_state=42)
low_dimensional_omv = tsne_omv.fit_transform(high_dimensional_omv)

high_dimensional_acc = np.array(clustered_acc_list)
tsne_acc = TSNE(n_components=1, random_state=42)
low_dimensional_acc = tsne_acc.fit_transform(high_dimensional_acc)
```

```python
curr_idx = 0
for i in range(len(plot_video_names)):
	video_name = plot_video_names[i]
	start = clip_amounts[i]
	end = clip_amounts[i+1]

	video_omv_cop1 = low_dimensional_omv[start:end, 0]
	video_acc_cop1 = low_dimensional_acc[start:end, 0]
	video_label = clustered_omv_labels[start:end]

	correlation_matrix = np.corrcoef(video_omv_cop1, video_acc_cop1)
	correlation_coefficient = correlation_matrix[0, 1]
	print(f'Video: {video_name} -> Corr: {correlation_coefficient}')

	plot_scatter_label(video_omv_cop1, video_acc_cop1, video_label, 'OMV', 'ACC', video_name)
```
